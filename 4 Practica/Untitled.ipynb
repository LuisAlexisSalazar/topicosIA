```python
import seaborn as sns
import pandas as pd 

df = pd.read_csv("Automobile_data.csv")

df.fillna(0)

df['price'] = (
    pd.to_numeric(df['price'],
                  errors='coerce')
      .fillna(0)
    )
df['engine-size'] = (
    pd.to_numeric(df['engine-size'],
                  errors='coerce')
      .fillna(0)
    )

sns.regplot(x='engine-size', y='price', data = df)
```

```python
sns.regplot(x='highway-mpg',y='price', data = df)
```

```python

df['peak-rpm'] = (
    pd.to_numeric(df['peak-rpm'],
                  errors='coerce')
      .fillna(0)
    )
```

```python
sns.regplot(x='peak-rpm',y='price',data=df)
```
