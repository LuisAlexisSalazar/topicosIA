```python
import numpy as np
import pandas as pd
```

```python
x_train = pd.read_csv('sample_data/train_ex1data1.csv')
x_test = pd.read_csv('sample_data/test_ex1data1.csv')
y_train = pd.read_csv('sample_data/y_train_ex1data1.csv')
y_test = pd.read_csv('sample_data/y_test_ex1data1.csv')
```

```python
def plot_sets():
    x_train.plot(title='x_train')
    y_train.plot(title='y_train')
    x_test.plot(title='x_test')
    y_test.plot(title='y_test')
    pd.concat([x_train, y_train], axis=1).plot(
        x='Num_Hab_Ciudad(Miles)', y='Rentabilidad_Ciudad(Miles)', kind='scatter', title='data_points'
    )
```

```python
plot_sets()
```

```python
x_mean = np.concatenate((x_train, x_test), axis=0).mean()
x_std = np.concatenate((x_train, x_test), axis=0).std()
y_mean = np.concatenate((y_train, y_test), axis=0).mean()
y_std = np.concatenate((y_train, y_test), axis=0).std()
```

```python
print(x_mean, x_std)
print(y_mean, y_std)
```

```python
x_train = (x_train - x_mean) / x_std
x_test = (x_test - x_mean) / x_std
y_train = (y_train - y_mean) / y_std
y_test = (y_test - y_mean) / y_std
```

```python
x_train.iloc[:,:]
```

```python
x_train = np.concatenate((np.ones([len(x_train), 1]), x_train), axis=1)
x_test = np.concatenate((np.ones([len(x_test), 1]), x_test), axis=1)
```

```python
theta = np.zeros(x_train.shape[1])
def predict(theta, data):
    return np.dot(data, theta)
```
